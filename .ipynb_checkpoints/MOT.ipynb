```python
#However, this doesn't provide a useful comparison. Better to use % pass rates.
```

```python
mot['Pass%'] = mot['FullTests_pass']/mot['FullTests_conducted']*100
mot['Retests%'] = mot['Retests_pass']/mot['Retests_conducted']*100
mot
```

```python
def roundup(value):
    result = round(value)
    return result
```

```python
roundup(83.400783)
```

```python
passrate = mot['Pass%']
passrate
```

```python
mot['passrate%'] = passrate.apply(roundup)
```

```python
retestrate = mot['Retests%']
retestrate
```

```python
mot['retestrate%'] = retestrate.apply(roundup)
```
